### 📘 Redacción del Problema
En el ámbito de la geometría básica y la programación, surge la necesidad de automatizar cálculos cotidianos. El problema consiste en diseñar un programa interactivo para estudiantes de preparatoria que permita calcular las dimensiones de un rectángulo.A diferencia de otras figuras, el rectángulo requiere la interacción con dos dimensiones independientes: la horizontal y la vertical. El reto es crear un sistema que, mediante un menú de opciones, reciba estos datos y devuelva resultados precisos, utilizando una estructura lógica de control (ciclos y condicionales) y modularizando el código a través de funciones para mantenerlo organizado y elegante.

### 📐 Fórmulas a utilizar
Para resolver los cálculos del rectángulo, aplicaremos las siguientes expresiones matemáticas:Área ($A$): Es el espacio total dentro de los límites de la figura.$$A = base \cdot altura$$Perímetro ($P$): Es la longitud total del contorno de la figura.$$P = 2 \cdot (base + altura)$$o también$$P = base + base + altura + altura$$

* Para que el programa funcione correctamente, debemos entender qué datos le estamos pidiendo al usuario:
* La Base ($b$): Representa el lado horizontal del rectángulo. Es la medida del ancho de la figura.La Altura ($h$):
* Representa el lado vertical del rectángulo. Es la medida de qué tan alto es el objeto.El Resultado:
* * El Área se expresa siempre en unidades cuadradas (como $cm^2$ o $m^2$), ya que representa una superficie.
* El Perímetro se expresa en unidades lineales (como $cm$ o $m$), ya que es una longitud.

```python
# ==========================================
# PROGRAMA: CALCULADORA DE RECTÁNGULOS 🟧
# programa realizado por Ramirez Azul NL = 44
# NIVEL: PRINCIPIANTE / PREPARATORIA
# ==========================================

# Función para calcular el área del rectángulo
def arearectangulo():
    print("\n--- 🟧 CALCULANDO ÁREA ---")
    base = float(input("📏 Ingresa la base del rectángulo: "))
    altura = float(input("📏 Ingresa la altura del rectángulo: "))
    area = base * altura
    print(f"✨ El área total es: {area:.2f} unidades²")

# Función para calcular el perímetro del rectángulo
def perimetrorectangulo():
    print("\n--- 📏 CALCULANDO PERÍMETRO ---")
    base = float(input("📐 Ingresa la base del rectángulo: "))
    altura = float(input("📐 Ingresa la altura del rectángulo: "))
    perimetro = 2 * (base + altura)
    print(f"✨ El perímetro total es: {perimetro:.2f} unidades")

# Función principal del menú
def menu():
    ejecutar = True

    while ejecutar:
        print("\n" + "═"*35)
        print("    🌟 MENÚ DE RECTÁNGULOS 🌟    ")
        print("═"*35)
        print("1️⃣  Calcular área rectángulo")
        print("2️⃣  Calcular perímetro rectángulo")
        print("3️⃣  Salir 🚪")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == "1":
            arearectangulo()
        elif opcion == "2":
            perimetrorectangulo()
        elif opcion == "3":
            print("\n👋 ¡Gracias por usar el programa! Hasta luego. ✨")
            ejecutar = False
        else:
            print("\n❌ Opción no válida. Por favor, elige 1, 2 o 3.")

# Llamada a la función menú para iniciar el programa
menu()
```
